```python
def remove_outliers(columns):
    for column in columns:
        q1=teams[column].quantile(.25)
        q3=teams[column].quantile(.75)
        iqr=q3-q1
        upper_boun=q3+1.5*iqr
        lower_bound=q1-1.5*iqr
        teams[column]=np.where(teams[column]>upper_boun,upper_boun,teams[column])
        teams[column]=np.where(teams[column]<lower_bound,lower_bound,teams[column])
```

```python
remove_outliers(["year","events","athletes","age","height","weight","medals","prev_medals","prev_3_medals"])
```

```python
def functionAnova(inpdata,targetvariable,categoricalpredictorlist):
    from scipy.stats import f_oneway
    selectedpredictors=[]
    print('###### AnovaResults ######\n')
    for predictor in categoricalpredictorlist:
        grouplists=inpdata.groupby[predictor](targetvariable).append(list)
        AnovaResults=f_oneway(*grouplists)
        if(AnovaResults[1]<0.05):
            print(predictor,'is correlated with',targetvariable,'/p_value:',AnovaResults[1])
            selectedpredictors.append(predictor)
        else:
            print(predictor,'is not correlated with',targetvariable,'/p_value:',AnovaResults[1])
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

def functionAnova(inpdata, targetvariable, categoricalpredictorlist):
    # Create a formula for the ANOVA model
    formula = f"{targetvariable} ~ " + " + ".join(categoricalpredictorlist)
    
    # Fit the model
    model = ols(formula, data=inpdata).fit()
    
    # Perform ANOVA
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    return anova_table

# Example usage
# Assuming 'bike' is your DataFrame
categoricalpredictorlist = ['team','country']
anova_results = functionAnova(inpdata=teams, targetvariable='events', categoricalpredictorlist=categoricalpredictorlist)

print(anova_results)
```
